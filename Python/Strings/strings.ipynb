# 1. Checking strings and if/else operation


```python
text = "The best thing in life are free"

print("free" in text)
print('expensice' in text)
print('expensive' not in text)

if 'expensive' not in text:
    print("No, 'expensive' is NOT present")
```

# 2. Slicing Strings


```python
a = 'Hello, World'

print(a[2:5])   # llo

# Slice from the beginning:
print(a[:5])    # Hello

# Slice to the End:
print(a[3:])    # llo, World

# Negative Indexingn
print(a[-5: -2])    # Wor
```

# 3. Modify Strings 


```python
a = "Hello, World"

# Upper Case
print(a.upper())

# Lower Case
print(a.lower())

b = "   Remove whitespace    "

# Remove Whitespace
print(b.strip())

# Remove Strings
print(a.replace("Hell", "Nic"))

c = "My, name, is, Emon, Rahman"
d = "My name is Emon Rahman"
# Split Strings
print(c.split( "," ))   # split for comma(,)
print(d.split( " " ))   #splited for whitespace (" ")
```

# 4. String Concatenation:


```python
a = 'hello'
b = 'world'

print(a + b)
print(a + " " + b)
```

# 5. Format Strings


```python
# 1. String Format

age = 23
text = "My age is " + str(age)

print(text)
```

```python
# 2. F-String/ Place-Holder

age = 23
text = f"My age is {age}"

print(text)
```

```python
# 3. Displaying price using 2 Decimals

price = 43
text = f"The price is {price:.2f} dollars"

print(text)
```

```python
# 4. Math Operation

price = 29
print(f"The price is {price* 5} dollars")
```

# 5. Escape Characteristics
    To insert characters that are illegal in a string use an escape character.

```python
# text = "We are "Vikings", from the North." -> Wrong
text = "We are \"Vikings\", from the North."
print(text)

```

```python
# 1. \' -> Single Quote
print('It\'s alright')

# 2. \\ -> Backslash
print("Insert one \\ (One backslash)")

# 3. \n -> new line

# 4. \r o -> Carriage return
text = "Hello\rWorld\rEmon"
print(text)

# 5. \t -> tab

# 6. \b -> backspace (Erases one character (backspace))
text = "hello \bworld"
print(text)

# 7. \f -> from feed 

# 8. \ooo -> Octal Value
print("\110\145\154\154\157") # "Hello"
 
# 9. \xhh -> Hex Value
print("\x48\x65\x6c\x6c\x6f") # "Hello"
```
